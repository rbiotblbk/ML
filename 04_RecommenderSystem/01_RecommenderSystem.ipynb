```python
import pandas as pd 
import numpy as np
```

# Load Data

```python
df_rating = pd.read_csv("./dataset/rating_final.csv")
df_geo = pd.read_csv("./dataset/geoplaces2.csv", encoding= "CP1252")
df_cuisine = pd.read_csv("./dataset/chefmozcuisine.csv")
 
```

# Constants

```python
TOP_N_PLACES = 10
```

# Clean Data

```python
# step over
```

```python
# 
places = df_geo[["placeID", "name", "city"]]
places.head()
```

# Get the Popular place

```python
# Group the places by rating --> take the average (mean) for each group(place)
rating =  pd.DataFrame(df_rating.groupby("placeID")["rating"].mean())

# Sort based on the rating
rating.sort_values("rating", ascending= False, inplace= True)

# Show the top 10x rated places
rating.head(TOP_N_PLACES)
```

# Chech how many rates for some restaurants (places)

for example: placeID : 132955 --> very high rated , but how many visits ??? --> 5x times

```python
df_rating.loc[df_rating["placeID"] == 132955]

# Alternative
df_rating.query("placeID == 132955")
```

To consider the visust count we need to know how many times (visited) are registed

```python
# Group the places by rating --> take the average (mean) for each group(place)
rating["rating_count"] =  pd.DataFrame(df_rating.groupby("placeID")["rating"].count())


# Sort based on the rating
rating.sort_values("rating_count", ascending= False, inplace= True)

# Show the top 10x most visited places
rating.head(TOP_N_PLACES)
```

# Check the information about the most popular (most reviewed place)

```python
top_popular_placeID = rating.head(1).index[0]


# Get the name of that placeID 
places.loc[places["placeID"] == top_popular_placeID]
```

```python
# Get the cuisine of that most visited(rated) place
df_cuisine.loc[df_cuisine["placeID"] == top_popular_placeID]
```

# The final answer is 


The most popular place is "Tortas Locas Hipocampo" which offers "Fast_Food" cuisne with average  rate score of "1,33"
